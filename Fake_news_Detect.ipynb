```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv('fake_data.csv')

x_train, x_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

vectorize = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
x_train_vec = vectorize.fit_transform(x_train)
x_test_vec = vectorize.transform(x_test)

model = LogisticRegression(max_iter=1000)
model.fit(x_train_vec, y_train)

y_pred = model.predict(x_test_vec)

print("Accuracy score:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
data_predict = [
    "Scientists discover a new species of bird in the Amazon rainforest",
    "Aliens have landed in the city park and are giving interviews"
]

vectorize_data = vectorize.transform(data_predict)

predictions = model.predict(vectorize_data)

for text, pred in zip(data_predict, predictions):
    print(f"Text: {text}\nPredictions: {pred}\n")
```
